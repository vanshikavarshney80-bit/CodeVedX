```python
import pandas as pd
import pickle
import nltk

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
news = [

# ---------- REAL NEWS ----------
"Government launches new education policy",
"India wins cricket world cup",
"Scientists discover new medicine",
"School opens new library",
"New technology improves farming",
"Hospital opens new emergency wing",
"University announces scholarship program",
"City builds new public park",
"Researchers develop faster computer chip",
"New railway line inaugurated",
"Government announces tax reduction",
"Doctors find effective dengue treatment",
"New airport terminal opens",
"Students win national science competition",
"State launches free health campaign",
"Rainfall improves crop production",
"Police rescue missing child safely",
"Electric buses introduced in the city",
"Company creates eco-friendly batteries",
"Engineers design energy-efficient building",

# ---------- FAKE NEWS ----------
"Aliens attacked Delhi yesterday",
"People become invisible after drinking tea",
"Moon is made of chocolate",
"Dinosaurs seen in New York",
"Ghost elected as Prime Minister",
"Cats start speaking English",
"Sun rises from the west today",
"Flying elephants spotted in Mumbai",
"Fish begin walking on roads",
"Trees started producing gold coins",
"Students become invisible after exam",
"Robots replace every teacher overnight",
"Time travel machine sold in local market",
"Earth has stopped rotating",
"Clouds made entirely of cotton candy",
"Dragon seen flying over the Himalayas",
"Zombie wins election in Europe",
"People live without food for 10 years",
"Cars now run only on water magically",
"Laptop grows on trees naturally"

]

label = [

"Real","Real","Real","Real","Real",
"Real","Real","Real","Real","Real",
"Real","Real","Real","Real","Real",
"Real","Real","Real","Real","Real",

"Fake","Fake","Fake","Fake","Fake",
"Fake","Fake","Fake","Fake","Fake",
"Fake","Fake","Fake","Fake","Fake",
"Fake","Fake","Fake","Fake","Fake"

]

data = pd.DataFrame({
    "News": news,
    "Label": label
})

print(data)
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(data["News"])

y = data["Label"]
```

```python
model = MultinomialNB()

model.fit(X,y)

print("Model Trained Successfully")
```

```python
pickle.dump(model,open("fake_news_model.pkl","wb"))

pickle.dump(vectorizer,open("vectorizer.pkl","wb"))

print("Model Saved Successfully")
```

```python
text = input("Enter News : ")

text_vector = vectorizer.transform([text])

prediction = model.predict(text_vector)

confidence = model.predict_proba(text_vector)

print("\nPrediction :",prediction[0])

print("Confidence :",round(max(confidence[0])*100,2),"%")
```

```python
loaded_model = pickle.load(open("fake_news_model.pkl","rb"))

loaded_vectorizer = pickle.load(open("vectorizer.pkl","rb"))

print("Model Loaded Successfully")
```

```python
news = input("Enter Another News : ")

vector = loaded_vectorizer.transform([news])

prediction = loaded_model.predict(vector)

confidence = loaded_model.predict_proba(vector)

print("Prediction :",prediction[0])

print("Confidence :",round(max(confidence[0])*100,2),"%")
```
