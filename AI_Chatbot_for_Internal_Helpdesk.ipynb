```python
import pandas as pd
import pickle

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
faq = {

"Question":[

"How do I reset my password?",
"How can I contact HR?",
"What are office timings?",
"Where is the attendance portal?",
"How to apply for leave?",
"How do I submit reimbursement?",
"Where can I download salary slips?",
"How do I access company email?",
"What is the IT support email?",
"How can I update my profile?",

"Password not working",
"I forgot my password",
"Need leave",
"Apply leave",
"Office timing",
"HR contact",
"Attendance issue",
"Salary slip",
"Company email",
"Profile update"

],

"Intent":[

"password",
"hr",
"office",
"attendance",
"leave",
"reimbursement",
"salary",
"email",
"it",
"profile",

"password",
"password",
"leave",
"leave",
"office",
"hr",
"attendance",
"salary",
"email",
"profile"

]

}

data = pd.DataFrame(faq)

print(data)
```

```python
vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(data["Question"])

y = data["Intent"]

model = MultinomialNB()

model.fit(X,y)

print("Chatbot Trained Successfully")
```

```python
pickle.dump(model,open("chatbot_model.pkl","wb"))
pickle.dump(vectorizer,open("vectorizer.pkl","wb"))
```

```python
reply = {

"password":"Please click on 'Forgot Password' to reset your password.",

"hr":"You can contact HR at hr@company.com",

"office":"Office timing is 9:00 AM to 6:00 PM.",

"attendance":"Attendance portal: https://attendance.company.com",

"leave":"Apply leave through the Employee Portal.",

"salary":"Salary slips are available in the HR Portal.",

"email":"Use Outlook to access company email.",

"it":"Contact IT Support: it@company.com",

"profile":"Update your profile in the Employee Portal.",

"reimbursement":"Submit reimbursement through Finance Portal."

}
```

```python
while True:

    user = input("\nYou : ")

    if user.lower()=="exit":
        break

    text = vectorizer.transform([user])

    intent = model.predict(text)[0]

    confidence = model.predict_proba(text)

    print("Bot :",reply[intent])

    print("Confidence :",round(max(confidence[0])*100,2),"%")
```
